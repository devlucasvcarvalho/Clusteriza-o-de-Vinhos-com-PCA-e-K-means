```python
# -*- coding: utf-8 -*-
"""
🍷 CLUSTERIZAÇÃO DE VINHOS COM PCA E K-MEANS
Google Colab Version
"""

# 1. INSTALAÇÃO E IMPORTAÇÃO DE BIBLIOTECAS
# ------------------------------------
print("📦 INSTALANDO E IMPORTANDO BIBLIOTECAS...")

!pip install -q plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, confusion_matrix

# Configurações de visualização
plt.style.use('default')
sns.set_palette("husl")
%matplotlib inline

print("✅ Bibliotecas carregadas com sucesso!")

# 2. CARREGAMENTO E EXPLORAÇÃO DOS DADOS
# ------------------------------------
print("\n" + "="*60)
print("📊 CARREGANDO DATASET DE VINHOS")
print("="*60)

# Carrega o dataset
wine_data = load_wine()
X = wine_data.data
y_true = wine_data.target
feature_names = wine_data.feature_names
target_names = wine_data.target_names

# Criar DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['classe_real'] = y_true
df['tipo_vinho'] = df['classe_real'].map({0: target_names[0], 1: target_names[1], 2: target_names[2]})

print(f"📈 INFORMAÇÕES DO DATASET:")
print(f"• Formato dos dados: {X.shape}")
print(f"• Número de características: {len(feature_names)}")
print(f"• Número de amostras: {len(X)}")
print(f"• Classes: {list(target_names)}")

# Visualização rápida dos dados
print("\n📋 PRIMEIRAS 5 AMOSTRAS:")
display(df.head())

print("\n📊 ESTATÍSTICAS DESCRITIVAS:")
display(df[feature_names].describe())

# 3. ANÁLISE EXPLORATÓRIA
# ------------------------------------
print("\n" + "="*60)
print("🔍 ANÁLISE EXPLORATÓRIA DOS DADOS")
print("="*60)

# Distribuição das classes
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = df['tipo_vinho'].value_counts()
plt.bar(class_counts.index, class_counts.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribuição das Classes de Vinho')
plt.xlabel('Tipo de Vinho')
plt.ylabel('Quantidade')
for i, count in enumerate(class_counts.values):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')

# Boxplot de algumas características
plt.subplot(1, 3, 2)
sample_features = feature_names[:4]  # Primeiras 4 características
df_sample = df[sample_features + ['tipo_vinho']]
df_melted = pd.melt(df_sample, id_vars=['tipo_vinho'], value_vars=sample_features)
sns.boxplot(data=df_melted, x='variable', y='value', hue='tipo_vinho')
plt.title('Distribuição por Característica e Tipo')
plt.xticks(rotation=45)

# Matriz de correlação
plt.subplot(1, 3, 3)
correlation_matrix = df[feature_names].corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Matriz de Correlação entre Variáveis')

plt.tight_layout()
plt.show()

# 4. PRÉ-PROCESSAMENTO
# ------------------------------------
print("\n" + "="*60)
print("🔧 PRÉ-PROCESSAMENTO DOS DADOS")
print("="*60)

# Escalonamento dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("✅ Dados escalonados com StandardScaler")
print(f"• Média após escalonamento: ~{np.mean(X_scaled):.2f}")
print(f"• Desvio padrão após escalonamento: ~{np.std(X_scaled):.2f}")

# 5. APLICAÇÃO DO PCA
# ------------------------------------
print("\n" + "="*60)
print("🎯 APLICAÇÃO DO PCA - REDUÇÃO DE DIMENSIONALIDADE")
print("="*60)

# PCA com 2 componentes para visualização
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# DataFrame com resultados do PCA
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Classe_Real'] = y_true
pca_df['Tipo_Vinho'] = pca_df['Classe_Real'].map({0: target_names[0], 1: target_names[1], 2: target_names[2]})

print("📊 RESULTADOS DO PCA:")
print(f"• Variância explicada por PC1: {pca.explained_variance_ratio_[0]:.3f} ({pca.explained_variance_ratio_[0]*100:.1f}%)")
print(f"• Variância explicada por PC2: {pca.explained_variance_ratio_[1]:.3f} ({pca.explained_variance_ratio_[1]*100:.1f}%)")
print(f"• Variância TOTAL explicada: {sum(pca.explained_variance_ratio_):.3f} ({sum(pca.explained_variance_ratio_)*100:.1f}%)")

# Visualização dos dados após PCA
plt.figure(figsize=(10, 6))
scatter = plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Classe_Real'],
                     cmap='viridis', s=60, alpha=0.7)
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.title('Visualização dos Dados após PCA\n(Coloração por Classe Real)')
plt.colorbar(scatter, label='Classe Real')
plt.grid(True, alpha=0.3)
plt.show()

# 6. MÉTODO DO COTOVELO PARA K IDEAL
# ------------------------------------
print("\n" + "="*60)
print("📈 DETERMINANDO NÚMERO IDEAL DE CLUSTERS")
print("="*60)

# Método do cotovelo
wcss = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, 'bo-', linewidth=2, markersize=8, alpha=0.7)
plt.axvline(x=3, color='red', linestyle='--', alpha=0.7, label='k=3 (conhecido)')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Método do Cotovelo - Número Ideal de Clusters')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("✅ Análise do cotovelo concluída - k=3 selecionado")

# 7. APLICAÇÃO DO K-MEANS
# ------------------------------------
print("\n" + "="*60)
print("🔍 APLICANDO ALGORITMO K-MEANS")
print("="*60)

# K-Means com k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_pca)

# Adicionar clusters ao DataFrame
pca_df['Cluster_KMeans'] = y_kmeans

# Métricas de avaliação
ari = adjusted_rand_score(y_true, y_kmeans)

print("📊 RESULTADOS DO K-MEANS:")
print(f"• Número de clusters: 3")
print(f"• Índice Ajustado Rand: {ari:.4f}")
print(f"• Coordenadas dos centróides:")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"  🎯 Cluster {i}: PC1={center[0]:.2f}, PC2={center[1]:.2f}")

# 8. VISUALIZAÇÃO COMPARATIVA
# ------------------------------------
print("\n" + "="*60)
print("📊 VISUALIZAÇÃO DOS RESULTADOS")
print("="*60)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico 1: Clusters do K-Means
scatter1 = axes[0].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster_KMeans'],
                          cmap='tab10', s=60, alpha=0.8)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
               c='red', s=200, marker='X', label='Centróides', edgecolors='black')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].set_title('Clusters do K-Means')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Classes Reais
scatter2 = axes[1].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Classe_Real'],
                          cmap='Set2', s=60, alpha=0.8)
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('Classes Reais')
axes[1].grid(True, alpha=0.3)

# Gráfico 3: Comparação lado a lado
for cluster in range(3):
    cluster_data = pca_df[pca_df['Cluster_KMeans'] == cluster]
    axes[2].scatter(cluster_data['PC1'], cluster_data['PC2'],
                   label=f'Cluster {cluster}', s=60, alpha=0.8)

axes[2].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
               c='black', s=200, marker='X', label='Centróides')
axes[2].set_xlabel('PC1')
axes[2].set_ylabel('PC2')
axes[2].set_title('Clusters (Sem Cores)')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 9. ANÁLISE DE PERFORMANCE
# ------------------------------------
print("\n" + "="*60)
print("📈 ANÁLISE DE PERFORMANCE")
print("="*60)

# Matriz de confusão
cm = confusion_matrix(y_true, y_kmeans)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cluster 0', 'Cluster 1', 'Cluster 2'],
            yticklabels=target_names)
plt.title('Matriz de Confusão: Clusters vs Classes Reais')
plt.xlabel('Clusters do K-Means')
plt.ylabel('Classes Reais')
plt.show()

# 10. VISUALIZAÇÃO INTERATIVA (OPCIONAL)
# ------------------------------------
print("\n" + "="*60)
print("🎨 VISUALIZAÇÃO INTERATIVA (Plotly)")
print("="*60)

try:
    fig = px.scatter(pca_df, x='PC1', y='PC2', color='Cluster_KMeans',
                     hover_data=['Tipo_Vinho'],
                     title='Visualização Interativa - Clusters K-Means',
                     labels={'Cluster_KMeans': 'Cluster'})
    fig.show()

    fig2 = px.scatter(pca_df, x='PC1', y='PC2', color='Tipo_Vinho',
                     title='Visualização Interativa - Classes Reais')
    fig2.show()
    print("✅ Visualizações interativas geradas!")
except Exception as e:
    print(f"⚠️  Visualização interativa não disponível: {e}")

# 11. RESUMO FINAL
# ------------------------------------
print("\n" + "="*60)
print("🎯 RESUMO E CONCLUSÕES")
print("="*60)

print(f"""
📊 RESULTADOS OBTIDOS:

1. PRÉ-PROCESSAMENTO:
   • Dataset com {X.shape[1]} características e {X.shape[0]} amostras
   • Dados escalonados com sucesso

2. REDUÇÃO DE DIMENSIONALIDADE (PCA):
   • 13D → 2D com {sum(pca.explained_variance_ratio_) * 100:.1f}% de variância mantida
   • Visualização eficiente dos dados

3. CLUSTERIZAÇÃO (K-MEANS):
   • 3 clusters identificados
   • Similaridade com classes reais: {ari:.1%}
   • Clusters bem separados e coerentes

💡 INSIGHTS:
   • As características químicas são bons discriminadores de tipos de vinho
   • PCA mostrou estrutura clara nos dados
   • K-Means encontrou grupos naturais compatíveis com a classificação real

✅ PROJETO CONCLUÍDO COM SUCESSO!
""")

print("="*60)
```
