```python
# -*- coding: utf-8 -*-
"""
üç∑ CLUSTERIZA√á√ÉO DE VINHOS COM PCA E K-MEANS
Google Colab Version
"""

# 1. INSTALA√á√ÉO E IMPORTA√á√ÉO DE BIBLIOTECAS
# ------------------------------------
print("üì¶ INSTALANDO E IMPORTANDO BIBLIOTECAS...")

!pip install -q plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, confusion_matrix

# Configura√ß√µes de visualiza√ß√£o
plt.style.use('default')
sns.set_palette("husl")
%matplotlib inline

print("‚úÖ Bibliotecas carregadas com sucesso!")

# 2. CARREGAMENTO E EXPLORA√á√ÉO DOS DADOS
# ------------------------------------
print("\n" + "="*60)
print("üìä CARREGANDO DATASET DE VINHOS")
print("="*60)

# Carrega o dataset
wine_data = load_wine()
X = wine_data.data
y_true = wine_data.target
feature_names = wine_data.feature_names
target_names = wine_data.target_names

# Criar DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['classe_real'] = y_true
df['tipo_vinho'] = df['classe_real'].map({0: target_names[0], 1: target_names[1], 2: target_names[2]})

print(f"üìà INFORMA√á√ïES DO DATASET:")
print(f"‚Ä¢ Formato dos dados: {X.shape}")
print(f"‚Ä¢ N√∫mero de caracter√≠sticas: {len(feature_names)}")
print(f"‚Ä¢ N√∫mero de amostras: {len(X)}")
print(f"‚Ä¢ Classes: {list(target_names)}")

# Visualiza√ß√£o r√°pida dos dados
print("\nüìã PRIMEIRAS 5 AMOSTRAS:")
display(df.head())

print("\nüìä ESTAT√çSTICAS DESCRITIVAS:")
display(df[feature_names].describe())

# 3. AN√ÅLISE EXPLORAT√ìRIA
# ------------------------------------
print("\n" + "="*60)
print("üîç AN√ÅLISE EXPLORAT√ìRIA DOS DADOS")
print("="*60)

# Distribui√ß√£o das classes
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
class_counts = df['tipo_vinho'].value_counts()
plt.bar(class_counts.index, class_counts.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribui√ß√£o das Classes de Vinho')
plt.xlabel('Tipo de Vinho')
plt.ylabel('Quantidade')
for i, count in enumerate(class_counts.values):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')

# Boxplot de algumas caracter√≠sticas
plt.subplot(1, 3, 2)
sample_features = feature_names[:4]  # Primeiras 4 caracter√≠sticas
df_sample = df[sample_features + ['tipo_vinho']]
df_melted = pd.melt(df_sample, id_vars=['tipo_vinho'], value_vars=sample_features)
sns.boxplot(data=df_melted, x='variable', y='value', hue='tipo_vinho')
plt.title('Distribui√ß√£o por Caracter√≠stica e Tipo')
plt.xticks(rotation=45)

# Matriz de correla√ß√£o
plt.subplot(1, 3, 3)
correlation_matrix = df[feature_names].corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Matriz de Correla√ß√£o entre Vari√°veis')

plt.tight_layout()
plt.show()

# 4. PR√â-PROCESSAMENTO
# ------------------------------------
print("\n" + "="*60)
print("üîß PR√â-PROCESSAMENTO DOS DADOS")
print("="*60)

# Escalonamento dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("‚úÖ Dados escalonados com StandardScaler")
print(f"‚Ä¢ M√©dia ap√≥s escalonamento: ~{np.mean(X_scaled):.2f}")
print(f"‚Ä¢ Desvio padr√£o ap√≥s escalonamento: ~{np.std(X_scaled):.2f}")

# 5. APLICA√á√ÉO DO PCA
# ------------------------------------
print("\n" + "="*60)
print("üéØ APLICA√á√ÉO DO PCA - REDU√á√ÉO DE DIMENSIONALIDADE")
print("="*60)

# PCA com 2 componentes para visualiza√ß√£o
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# DataFrame com resultados do PCA
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Classe_Real'] = y_true
pca_df['Tipo_Vinho'] = pca_df['Classe_Real'].map({0: target_names[0], 1: target_names[1], 2: target_names[2]})

print("üìä RESULTADOS DO PCA:")
print(f"‚Ä¢ Vari√¢ncia explicada por PC1: {pca.explained_variance_ratio_[0]:.3f} ({pca.explained_variance_ratio_[0]*100:.1f}%)")
print(f"‚Ä¢ Vari√¢ncia explicada por PC2: {pca.explained_variance_ratio_[1]:.3f} ({pca.explained_variance_ratio_[1]*100:.1f}%)")
print(f"‚Ä¢ Vari√¢ncia TOTAL explicada: {sum(pca.explained_variance_ratio_):.3f} ({sum(pca.explained_variance_ratio_)*100:.1f}%)")

# Visualiza√ß√£o dos dados ap√≥s PCA
plt.figure(figsize=(10, 6))
scatter = plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Classe_Real'],
                     cmap='viridis', s=60, alpha=0.7)
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.title('Visualiza√ß√£o dos Dados ap√≥s PCA\n(Colora√ß√£o por Classe Real)')
plt.colorbar(scatter, label='Classe Real')
plt.grid(True, alpha=0.3)
plt.show()

# 6. M√âTODO DO COTOVELO PARA K IDEAL
# ------------------------------------
print("\n" + "="*60)
print("üìà DETERMINANDO N√öMERO IDEAL DE CLUSTERS")
print("="*60)

# M√©todo do cotovelo
wcss = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, 'bo-', linewidth=2, markersize=8, alpha=0.7)
plt.axvline(x=3, color='red', linestyle='--', alpha=0.7, label='k=3 (conhecido)')
plt.xlabel('N√∫mero de Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('M√©todo do Cotovelo - N√∫mero Ideal de Clusters')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("‚úÖ An√°lise do cotovelo conclu√≠da - k=3 selecionado")

# 7. APLICA√á√ÉO DO K-MEANS
# ------------------------------------
print("\n" + "="*60)
print("üîç APLICANDO ALGORITMO K-MEANS")
print("="*60)

# K-Means com k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_pca)

# Adicionar clusters ao DataFrame
pca_df['Cluster_KMeans'] = y_kmeans

# M√©tricas de avalia√ß√£o
ari = adjusted_rand_score(y_true, y_kmeans)

print("üìä RESULTADOS DO K-MEANS:")
print(f"‚Ä¢ N√∫mero de clusters: 3")
print(f"‚Ä¢ √çndice Ajustado Rand: {ari:.4f}")
print(f"‚Ä¢ Coordenadas dos centr√≥ides:")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"  üéØ Cluster {i}: PC1={center[0]:.2f}, PC2={center[1]:.2f}")

# 8. VISUALIZA√á√ÉO COMPARATIVA
# ------------------------------------
print("\n" + "="*60)
print("üìä VISUALIZA√á√ÉO DOS RESULTADOS")
print("="*60)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Gr√°fico 1: Clusters do K-Means
scatter1 = axes[0].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster_KMeans'],
                          cmap='tab10', s=60, alpha=0.8)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
               c='red', s=200, marker='X', label='Centr√≥ides', edgecolors='black')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].set_title('Clusters do K-Means')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Gr√°fico 2: Classes Reais
scatter2 = axes[1].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Classe_Real'],
                          cmap='Set2', s=60, alpha=0.8)
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('Classes Reais')
axes[1].grid(True, alpha=0.3)

# Gr√°fico 3: Compara√ß√£o lado a lado
for cluster in range(3):
    cluster_data = pca_df[pca_df['Cluster_KMeans'] == cluster]
    axes[2].scatter(cluster_data['PC1'], cluster_data['PC2'],
                   label=f'Cluster {cluster}', s=60, alpha=0.8)

axes[2].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
               c='black', s=200, marker='X', label='Centr√≥ides')
axes[2].set_xlabel('PC1')
axes[2].set_ylabel('PC2')
axes[2].set_title('Clusters (Sem Cores)')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 9. AN√ÅLISE DE PERFORMANCE
# ------------------------------------
print("\n" + "="*60)
print("üìà AN√ÅLISE DE PERFORMANCE")
print("="*60)

# Matriz de confus√£o
cm = confusion_matrix(y_true, y_kmeans)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cluster 0', 'Cluster 1', 'Cluster 2'],
            yticklabels=target_names)
plt.title('Matriz de Confus√£o: Clusters vs Classes Reais')
plt.xlabel('Clusters do K-Means')
plt.ylabel('Classes Reais')
plt.show()

# 10. VISUALIZA√á√ÉO INTERATIVA (OPCIONAL)
# ------------------------------------
print("\n" + "="*60)
print("üé® VISUALIZA√á√ÉO INTERATIVA (Plotly)")
print("="*60)

try:
    fig = px.scatter(pca_df, x='PC1', y='PC2', color='Cluster_KMeans',
                     hover_data=['Tipo_Vinho'],
                     title='Visualiza√ß√£o Interativa - Clusters K-Means',
                     labels={'Cluster_KMeans': 'Cluster'})
    fig.show()

    fig2 = px.scatter(pca_df, x='PC1', y='PC2', color='Tipo_Vinho',
                     title='Visualiza√ß√£o Interativa - Classes Reais')
    fig2.show()
    print("‚úÖ Visualiza√ß√µes interativas geradas!")
except Exception as e:
    print(f"‚ö†Ô∏è  Visualiza√ß√£o interativa n√£o dispon√≠vel: {e}")

# 11. RESUMO FINAL
# ------------------------------------
print("\n" + "="*60)
print("üéØ RESUMO E CONCLUS√ïES")
print("="*60)

print(f"""
üìä RESULTADOS OBTIDOS:

1. PR√â-PROCESSAMENTO:
   ‚Ä¢ Dataset com {X.shape[1]} caracter√≠sticas e {X.shape[0]} amostras
   ‚Ä¢ Dados escalonados com sucesso

2. REDU√á√ÉO DE DIMENSIONALIDADE (PCA):
   ‚Ä¢ 13D ‚Üí 2D com {sum(pca.explained_variance_ratio_) * 100:.1f}% de vari√¢ncia mantida
   ‚Ä¢ Visualiza√ß√£o eficiente dos dados

3. CLUSTERIZA√á√ÉO (K-MEANS):
   ‚Ä¢ 3 clusters identificados
   ‚Ä¢ Similaridade com classes reais: {ari:.1%}
   ‚Ä¢ Clusters bem separados e coerentes

üí° INSIGHTS:
   ‚Ä¢ As caracter√≠sticas qu√≠micas s√£o bons discriminadores de tipos de vinho
   ‚Ä¢ PCA mostrou estrutura clara nos dados
   ‚Ä¢ K-Means encontrou grupos naturais compat√≠veis com a classifica√ß√£o real

‚úÖ PROJETO CONCLU√çDO COM SUCESSO!
""")

print("="*60)
```
